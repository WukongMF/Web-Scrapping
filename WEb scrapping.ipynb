```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Image, display

# URL de la página que deseas scrapear
url = 'https://www.sears.com.mx/producto/204461/blusa-shifon-con-corbatin-manga-larga-michel-paulini-para-mujer'  # Reemplaza esto con la URL de la página que desees scrapear

# Realizamos una solicitud HTTP para obtener el contenido de la página
response = requests.get(url)

# Verificamos si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Analizamos el contenido de la página con BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encuentra todas las etiquetas 'img' en la página
    img_tags = soup.find_all('img')

    for i in [1, 6]:  # Índices 0 y 5 corresponden a la primera y sexta imagen
        img_tag = img_tags[i]
        img_url = img_tag.get('src')
        if img_url:
            img_url = img_url if img_url.startswith('http') else url + img_url
            display(Image(url=img_url))
else:
    print('La solicitud HTTP no fue exitosa. Código de estado:', response.status_code)





```

```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Image, display
import ipywidgets as widgets
from IPython.display import display as ipydisplay

# URL de la página que deseas scrapear en Shein
url = 'https://www.shein.com/'  # Reemplaza esto con la URL de la página que desees scrapear en Shein

# Realizamos una solicitud HTTP para obtener el contenido de la página
response = requests.get(url)

# Verificamos si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Analizamos el contenido de la página con BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encuentra todas las etiquetas 'img' en la página
    img_tags = soup.find_all('img')

    # Crear un widget para seleccionar cuántos elementos mostrar
    num_elements_widget = widgets.IntSlider(value=1, min=1, max=max(1, len(img_tags)), description='Número de elementos:')

    def show_images(num_elements):
        if num_elements <= len(img_tags):
            for i in range(num_elements):
                img_tag = img_tags[i]
                img_url = img_tag.get('src')
                if img_url:
                    img_url = img_url if img_url.startswith('http') else url + img_url
                    display(Image(url=img_url))
        else:
            print(f'Solo se encontraron {len(img_tags)} imágenes en la página.')

    widgets.interactive(show_images, num_elements=num_elements_widget)
    ipydisplay(num_elements_widget)

else:
    print('La solicitud HTTP no fue exitosa. Código de estado:', response.status_code)
```
